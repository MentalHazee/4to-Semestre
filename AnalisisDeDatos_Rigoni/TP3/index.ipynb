### Ejercicio 1: Slicing y Sub-matrices

```python
import numpy as np

#1
matriz = np.arange(1, 26).reshape(5, 5)
print(matriz)

#2
subMatriz = matriz[1:4, 1:4]
print("\n", subMatriz)

#3
ultima_columna = matriz[:, -1]
print("\n", ultima_columna)
```

### Ejercicio 2: Estadísticas con Ejex (Axis)

```python
import numpy as np

ventas = np.array([
    [1500, 2000, 1800, 2200],
    [1100, 1300, 1200, 1500],
    [3000, 3100, 2900, 3500]
])

#1 
total = np.sum(ventas)
print(total)

#2
promedio_semanal = np.mean(ventas, axis=0)
print("\n", promedio_semanal)

#3
max_por_sucursal = np.max(ventas, axis=1)
print("\n", max_por_sucursal)
```

### Ejercicio 3: El poder de la Máscara Booleana

```python
import numpy as np

#1
arr = np.arange(1, 21)
print(arr)

#2
mascara =  arr % 2 == 0
print("\n", mascara)

#3
pares = arr[mascara]
print("\n", pares)
```

### Ejercicio 4: Condiciones Múltiples (El Analista de RRHH)

```python
import numpy as np

edades = np.array([22, 55, 34, 19, 45, 61, 28, 30, 42])

#1
filtro = (edades >= 30) & (edades <= 50)
resultado = edades[filtro]

print(resultado)

#2
filtro_extremos = (edades < 25) | (edades > 55)
resultado_extremos = edades[filtro_extremos]

print("\n", resultado_extremos)
```

### Ejercicio 5: Modificación Condicional

### Ejercicio 6: Desafío Integrador "El Profesor Robot" 

```python
import numpy as np

#1
calificaciones = np.random.randint(1, 11, 50)
print(calificaciones)

#2
promedio = np.mean(calificaciones)
print("\nPromedio:", promedio)

#3
desaprobados = calificaciones[calificaciones < 4]
print("\nNotas desaprobadas:", desaprobados)

cantidad_desaprobados = desaprobados.size
print("\nHay", cantidad_desaprobados, "desaprobados en total")

#4
calificaciones[calificaciones < 4] = 4

#5
print("\nNotas finales:", calificaciones)
```
