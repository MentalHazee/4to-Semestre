### Ejercicio 1: Creación del entorno de datos

```python
import csv
datos_tienda = [
 {"ID": 1001, "Producto": "Laptop Gamer", "Categoria": "Computacion", "Precio": 1200000, "Stock": 15},
 {"ID": 1002, "Producto": "Mouse Inalambrico", "Categoria": "Accesorios", "Precio": 25000, "Stock": 100},
 {"ID": 1003, "Producto": "Teclado Mecanico", "Categoria": "Accesorios", "Precio": 85000, "Stock": 40},
 {"ID": 1004, "Producto": "Monitor 27 Pulgadas", "Categoria": "Computacion", "Precio": 350000, "Stock":
25},
 {"ID": 1005, "Producto": "Auriculares Bluetooth", "Categoria": "Audio", "Precio": 60000, "Stock": ""}, 
 
 #Faltante intencional
 {"ID": 1006, "Producto": "Silla Ergonomica", "Categoria": "Muebles", "Precio": 250000, "Stock": 10}
]
with open('productos.csv', 'w', newline='', encoding='utf-8') as archivo:
 columnas = ["ID", "Producto", "Categoria", "Precio", "Stock"]
 writer = csv.DictWriter(archivo, fieldnames=columnas)
 writer.writeheader()
 writer.writerows(datos_tienda)
print("Archivo 'productos.csv' creado exitosamente.")
```
