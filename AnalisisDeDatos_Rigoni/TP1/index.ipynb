## Ejercicio 4

```python
ventas = [
    {"ticket": 1, "monto": 5000, "metodo": "Efectivo"},
    {"ticket": 2, "monto": 12000, "metodo": "Tarjeta"},
    {"ticket": 3, "monto": 8500, "metodo": "Efectivo"},
    {"ticket": 4, "monto": 3000, "metodo": "MercadoPago"}
]

total_dia = 0
total_efectivo = 0

for venta in ventas:
    total_dia += venta["monto"]
    
    if venta["metodo"] == "Efectivo":
        total_efectivo += venta["monto"]

print(f"Total recaudado en el día: ${total_dia}")
print(f"Total recaudado en efectivo: ${total_efectivo}")
```
